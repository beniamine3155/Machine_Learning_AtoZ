```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os 

```

```python
iris = sns.load_dataset('iris') # load the iris dataset from seaborn this is built in dataset
iris.head()
```

```python
iris.describe() # get the summary of the dataset
```

## Line plot

```python
plt.rcParams['figure.figsize'] = [8, 4] # rcParams is a dictionary that stores the default settings for matplotlib plots 
plt.rcParams['figure.dpi'] = 300 
iris.plot()
```

```python
iris.plot(logy=True, logx=True) # plot the dataset with log scale on x and y axis, log scale is used to represent the data in a more readable way when the data is very large
```

## Bar Plot

```python
df = iris.iloc[0][0:4] # [0:4] means we are selecting the first row and first 4 columns of the dataset
df
```

```python
df.plot(kind='bar') # plot the data in bar chart
```

```python

df.plot.bar( color='g', grid=True) # another way to plot the data in bar chart
plt.show()
```

## Stacked bar plot

```python
df = iris.head()
df.plot.bar(stacked=True, ylim=(0,12)) # stacked bar chart is used to represent the data in a more readable way when the data is very large
plt.legend(bbox_to_anchor=(1.0, 1.0)) # set the legend position
plt.show()
```

```python
df.plot.barh(stacked=True, ylim=(0,12)) # stacked bar chart is used to represent the data in a more readable way when the data is very large
plt.legend(bbox_to_anchor=(1.0, 1.0)) # set the legend position
plt.show()
```

## Histogram

```python
iris.plot.hist(bins=20 , alpha=0.7) # plot the histogram of the dataset with 20 bins and alpha value 0.7 , alpha value is used to set the transparency of the plot
```

```python
iris['petal_length'].plot(kind='hist', bins=20, orientation='horizontal', alpha=0.7) 
```

```python
iris.hist() # plot the histogram of the datset with default settings 
plt.show()
```

## Box Plot

```python
iris.plot.box(color = dict(), patch_artist=True, vert=False) # box plot is used to represent the data in a more readable way when the data is very large, box plot identifies the outliers int the data also labest the data in 25th, 50th and 75th percentile
```

## Area and Scatter Plot

```python
iris.plot.area() # area plot describe the data in a more readable way when the data is very large 
```

```python
iris.plot.scatter(y='sepal_length', x='sepal_width') # when we have two variables and we want to see the relationship between them we use scatter plot
```

```python
iris[iris['species']=='setosa'].plot.scatter(y='sepal_length', x='sepal_width')
```

```python
iris['species'].unique()
```

```python
iris[iris['species']=='versicolor'].plot.scatter(y='sepal_length', x='sepal_width')
```

```python
iris[iris['species']=='versicolor'].plot.scatter(y='sepal_length', x='sepal_width' , c='petal_length') # c is used to color the data points based on the petal length
```
