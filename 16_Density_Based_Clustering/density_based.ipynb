### Density Based Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import  make_moons
```

```python
X, y = make_moons(n_samples=200, noise=0.1, random_state=0 )
# n_samples = 200; 200 data points for each class
# noise = 0.1; standard deviation of Gaussian noise added to the data that helps separate the two classes, 0.1 indicates 10% of the data points are randomly distributed
# random_state = 0; random seed for data generation
```

```python
X, y
```

```python
plt.scatter(X[:, 0], X[:, 1]) # X[:, 0] is the first column of X, X[:, 1] is the second column of X
plt.show()
```

### DBSCAN Coding

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
labels = kmeans.labels_

# scatter plot of the data points and color the points according to the cluster labels
plt.scatter(X[:, 0], X[:, 1], c=labels) # c=labels; color the points according to the cluster labelsf, cmap
plt.show()
```

```python
# DBSCAN
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
# eps = 0.2; maximum distance between two samples for one to be considered as in the neighborhood of the other
# min_samples = 5; number of samples in a neighborhood for a point to be considered as a core point
# metric = 'euclidean'; the distance metric used to calculate the distance between two points
clusters = dbscan.fit_predict(X)

# scatter plot
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.show()
```
