## IMDB Movie Data Analysis

```python
movies = pd.read_csv('IMDB-Movie-Data.csv', nrows=50)
movies.head()
```

```python
movies.columns
```

```python
movies.isnull().sum() # returns the number of missing values in each column
```

```python
movies.fillna(0, inplace=True) # fill all NaN values with 0
```

```python
plt.figure(figsize=(12, 6)) # 12 inches wide and 6 inches tall
plt.plot(movies['Rank'], movies['Revenue (Millions)']) 
plt.show() # display the plot
```

```python
idx = movies['Revenue (Millions)'].idxmax() # return the index of the maximum value
movies.iloc[idx] # return the row with the maximum value
```

```python


plt.figure(figsize=(12,6))
plt.plot(movies['Rank'], movies['Runtime (Minutes)'])
plt.show()
```

## Linestyling and Putting labels

```python
plt.figure(figsize=(10, 3))
plt.plot(movies['Rank'], movies['Rating'], color='g', linestyle='--', linewidth=2, marker='o', markersize=8, markerfacecolor='r') # g = green color, -- = dashed line, linewidth = 2, o = circle marker
plt.xlabel('Rank') 
plt.ylabel('Rating')
plt.title('Rank vs Rating')
plt.show()
```

## Scatter, Bar, Histogram plot

```python
plt.figure(figsize=(10, 3))
plt.scatter(movies['Rank'], movies['Rating'])

plt.figure(figsize=(10, 3))
plt.bar(movies['Rank'], movies['Rating'], color='g' )

plt.figure(figsize=(5, 10))
plt.barh(movies['Rank'], movies['Rating'], color='orange' )
```

```python
plt.figure(figsize=(12, 5))
plt.plot(movies['Rank'], movies['Rating'], color='y', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
plt.bar(movies['Rank'], movies['Rating'], color='g' )
```

```python
plt.figure(figsize=(10, 5))
plt.hist(movies['Rating'], color='orange', alpha=0.7, bins=15 ) # alpha is the transparency ot the color
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot(movies['Rating'], patch_artist=True) # patch_artish fills the boxplot with color
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1) # 1 row and 2 columns and this is the first plot 
plt.bar(movies['Rank'], movies['Rating'], color='g')

plt.subplot(1, 2, 2) # 1 row and 2 columns and this is the second plot 
plt.hist(movies['Rating'], bins=20, color='g')
plt.show()
```

```python
# 2x2 subplots
plt.figure(figsize=(12,12))
# 1---------------------------------
plt.subplot(2, 2, 1) 
plt.plot(movies['Rank'], movies['Rating'], color='g', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
plt.xlabel('Rank')
plt.ylabel('Rating')
plt.title('Rank vs Rating')

# 2---------------------------------
plt.subplot(2, 2, 2)
plt.plot(movies['Rank'], movies['Revenue (Millions)'], color='y', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
plt.xlabel('Rank')
plt.ylabel('Revenue')
plt.title('Rank vs Revenue')

# 3---------------------------------
plt.subplot(2, 2, 3)
plt.plot(movies['Rank'], movies['Votes'], color='b', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
plt.xlabel('Rank')
plt.ylabel('Votes')
plt.title('Rank vs Votes')

# 4---------------------------------
plt.subplot(2, 2, 4)
plt.plot(movies['Rank'], movies['Metascore'], color='k', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
plt.xlabel('Rank')
plt.ylabel('Metascore')
plt.title('Rank vs Metascore')

plt.tight_layout() # adjust the subplots to fit into the figure area
```

```python
# 2 x 2 subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8)) # fig here is the figure object, ax is the axes object , fig is the canvas means the whole area where we can draw the plots, ax is the area where we can draw the plots
ax[0][0].plot(movies['Rank'], movies['Rating'], color='g', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
ax[0][0].set_xlabel('Rank')
ax[0][0].set_ylabel('Rating')
ax[0][0].set_title('Rank vs Rating')

ax[1][1].plot(movies['Rank'], movies['Votes'], color='k', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
ax[1][1].set_xlabel('Rank')
ax[1][1].set_ylabel('Votes')
ax[1][1].set_title('Rank vs Votes')

plt.tight_layout()
```

## Creating a zoom sub figure of figure

```python
fig = plt.figure(figsize=(10,3))
ax1 = fig.add_axes([0, 0, 1, 1]) # main axes in the figure with 100% width and height
ax2 = fig.add_axes([0.1, 0.1, 0.2, 0.4]) # inset axes in the figure with 20% width and 40% height
```

```python
fig = plt.figure(figsize=(10,3))
ax1 = fig.add_axes([0, 0, 1, 1]) 
ax2 = fig.add_axes([0.1, 0.1, 0.2, 0.4])
ax1.plot(movies['Rank'], movies['Rating'], color='g', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
ax1.set_xlabel('Rank')
ax1.set_ylabel('Rating')
ax1.set_title('Rank vs Rating')

## plot zoom subplot
ax2.plot(movies['Rank'], movies['Rating'], color='g', linestyle='-', linewidth=2, marker='o', markersize=8, markerfacecolor='r')
ax2.set_xlim(20, 25) # set the x-axis limits to 20 and 25
ax2.set_ylim(4, 8) # set the y-axis limits to 4 and 8
plt.show()
```

## xlim and ylim, legend, grid, xticks , yticks

```python
movies.head(1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(movies['Rank'], movies['Rating'], color='g', label="Rating")
plt.plot(movies['Rank'], movies['Rating'] + 1, color ='b', label = "Rating + 1")

plt.xlim(0, 30)
plt.ylim(4, 10)

plt.legend() # add the labels to the plot like rating and rating + 1

plt.grid() # grid() function adds grid lines to the plot
plt.xticks(np.arange(0, 30, 4)) # set the x-axis ticks 0 to 30 with an interval of 4
plt.yticks(np.arange(4, 10, 1)) # set the y-axis ticks 4 to 10 with an interval of 1
plt.show()
```

## Pie Chart

```python
df = pd.read_csv('quarters.csv')
df.head(5)
```

```python
plt.figure(figsize=(5, 5))
plt.pie(df['Q1'], labels=df['Salesman'], autopct='%1.1f%%', startangle=180 ) # %1.1f%% display the percentage with 1 decimal place, startangle rotates the pie chart by 180 degrees
plt.title('Revenue by Quarters')
plt.savefig('piechart.png', dpi = 300) # save the plot as a png file with 300 dpi(dots per inch)
plt.show()
```
